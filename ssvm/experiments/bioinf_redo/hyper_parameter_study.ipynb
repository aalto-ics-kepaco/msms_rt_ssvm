```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
```

## CASMI - NEGATIVE

```python
IDIR = "logs_triton/version_01a/ds=CASMI__ion=negative/"
```

```python
df = pd.concat(
    [
        pd.read_csv(os.path.join(IDIR, "grid_search_results__spl=%03d.tsv" % idx), sep="\t").assign(sample=idx) 
        for idx in range(50)
    ]
)

ax = sns.pointplot(x="C", y="ndcg_ohc", data=df)
ax.grid(axis="y")
ax.set_xlabel("Regularization parameter C")
ax.set_ylabel("NDCG on Test Set")
ax.set_title("Normalized Discounted Cumulative Gain (NDCG)")
```

- reduce search grid to [1, 4, 16, 32, 64]

```python
df = pd.concat(
    [
        pd.read_csv(os.path.join(IDIR, "topk__spl=%03d.tsv" % idx), sep="\t").assign(sample=idx) 
        for idx in range(50)
    ]
)
```

```python
ax = sns.pointplot(x="k", y="top_acc_perc", data=df[df["k"] <= 20])
ax.grid(axis="y")
ax.set_xlabel("k")
ax.set_ylabel("Top-k Accuracy (%)")
```

```python
df.loc[df["k"].isin([1, 5, 10, 20]), ["k", "top_acc_perc"]].groupby("k").mean()
```
